# Face Image Processing using OpenCV

### Objective
In this project, we perform basic image processing techniques on facial images.  
The goal is to understand how images are enhanced, edges are detected, and features are extracted for further analysis.
### Steps Overview
We will perform the following steps:
1. Load and display the dataset  
2. Enhance the images (brightness & contrast)  
3. Detect edges  
4. Smooth and denoise the images  
5. Extract image features  
6. Match features between two images  
7. Save and visualize processed results

###  Step 1: Load, Display, and Resize Images

```python
import os
import cv2

#Load Images
base_dir = r"C:\Users\Administrator\Desktop\Face DataSet"
faces_dir = os.path.join(base_dir, "faces_01", "Train")

#Get all image file names
image_files = [f for f in os.listdir(faces_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]
print(" Total images found:", len(image_files))
```

```python
import matplotlib.pyplot as plt

#Display a sample image
sample_path = os.path.join(faces_dir, image_files[0])
img = cv2.imread(sample_path)

if img is not None:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title("Sample Face Image")
    plt.axis("off")
    plt.show()
else:
    print(" Image not loaded. Check file path.")
```

#### Resizing to 128×128 ensures all images are uniform, faster to process, and still detailed enough for facial feature detection and clustering

```python
#Resize all images to 128x128
resized_images = []

for img_name in image_files:
    img_path = os.path.join(faces_dir, img_name)
    img = cv2.imread(img_path)
    if img is not None:
        img_resized = cv2.resize(img, (128, 128))
        resized_images.append(img_resized)

# Display one resized image
plt.imshow(cv2.cvtColor(resized_images[0], cv2.COLOR_BGR2RGB))
plt.title("Resized Image (128x128)")
plt.axis("off")
plt.show()

print(" Total images resized:", len(resized_images))
```

#### Converting to grayscale removes color noise, speeds up processing, and helps algorithms focus on key visual patterns like shapes, edges, and textures.

```python
# Convert resized images to grayscale
gray_images = [cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) for img in resized_images]

# Show a sample grayscale image
plt.imshow(gray_images[0], cmap='gray')
plt.title("Sample Grayscale Image")
plt.axis('off')
plt.show()

print(" Total grayscale images:", len(gray_images))
```

##### Histogram Equalization enhances image contrast by spreading out pixel intensity values — making dark areas brighter and bright areas clearer.

```python
# Enhance image contrast using Histogram Equalization
enhanced_images = [cv2.equalizeHist(img) for img in gray_images]

# Show before vs after enhancement
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(gray_images[0], cmap='gray')
plt.title("Original Grayscale")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(enhanced_images[0], cmap='gray')
plt.title("Enhanced Image")
plt.axis('off')

plt.show()
```

##### Canny edge detection is important because it helps identify the boundaries and outlines of objects in an image, making it easier for the computer to recognize shapes, detect features, and separate different regions during further image analysis.

```python
# Detect edges using the Canny Edge Detector
edges = [cv2.Canny(img, 100, 200) for img in enhanced_images]

# Show example result
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(enhanced_images[0], cmap='gray')
plt.title("Enhanced Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(edges[0], cmap='gray')
plt.title("Edges Detected")
plt.axis('off')

plt.show()
```

##### Applying Gaussian Blur reduces noise and unwanted details in an image, helping to make edges and features more consistent and improving the accuracy of later processing steps like edge detection or feature extraction.

```python
# Apply Gaussian Blur to smooth the images
smoothed_images = [cv2.GaussianBlur(img, (5,5), 0) for img in enhanced_images]

# Show before and after smoothing
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(enhanced_images[0], cmap='gray')
plt.title("Before Smoothing")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(smoothed_images[0], cmap='gray')
plt.title("After Smoothing")
plt.axis('off')

plt.show()
```

##### Initializing ORB detector helps identify keypoints and compute their descriptors, which represent unique and identifiable features in an image.

```python
# Initialize ORB detector
orb = cv2.ORB_create()

# Detect and compute keypoints and descriptors for one image
keypoints, descriptors = orb.detectAndCompute(smoothed_images[0], None)

# Draw keypoints on the image
img_with_keypoints = cv2.drawKeypoints(smoothed_images[0], keypoints, None, color=(0,255,0), flags=0)

# Show result
plt.figure(figsize=(6,6))
plt.imshow(img_with_keypoints, cmap='gray')
plt.title("Feature Points (ORB)")
plt.axis('off')
plt.show()

print(f"Number of keypoints detected: {len(keypoints)}")
```

##### Feature matching is done to test the effectiveness of feature detection and matching after preprocessing, showing that even after transformations, the key features of faces remain recognizable.

```python
# Take two sample images
img1 = smoothed_images[0]
img2 = smoothed_images[1]

# Initialize ORB detector
orb = cv2.ORB_create()

# Detect keypoints and descriptors for both images
kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2, None)

# Create Brute-Force Matcher
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Match descriptors
matches = bf.match(des1, des2)

# Sort matches by distance (smaller distance = better match)
matches = sorted(matches, key=lambda x: x.distance)

# Draw first 25 matches
matched_img = cv2.drawMatches(img1, kp1, img2, kp2, matches[:25], None, flags=2)

# Show result
plt.figure(figsize=(12,6))
plt.imshow(matched_img)
plt.title("Feature Matching between Two Images")
plt.axis('off')
plt.show()

print(f"Total matches found: {len(matches)}")
```

#### Now storing and documenting the results of your image processing pipeline for future use or demonstration.

```python
import os

# Create output folder if not exists
output_dir = os.path.join(base_dir, "Processed_Results")
os.makedirs(output_dir, exist_ok=True)

# Save a few processed images
for i, img in enumerate(smoothed_images[:5]):
    save_path = os.path.join(output_dir, f"processed_{i+1}.jpg")
    cv2.imwrite(save_path, img)

print(f"✅ Saved {len(smoothed_images[:5])} processed images in: {output_dir}")
```
